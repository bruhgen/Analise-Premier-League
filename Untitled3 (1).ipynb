```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
Base_Dados = pd.read_csv('results.csv')
Base_Dados.head()
```

```python
plt.figure(figsize =(14,5))
plt.title('Quantidade de Gols Premier League Temporadas 2006 a 2018', loc='left', fontsize = 14)
sns.barplot(data = Base_Dados, x = 'home_goals', y = 'season', ci= None, color ='#69b3a2', estimator = sum );
plt.ylabel('Temporadas');
```

```python
plt.figure(figsize =(14,5))
plt.title('Quantidade de Gols Premier League Temporadas 2006 a 2018', loc='left', fontsize = 14)
sns.barplot(data = Base_Dados, x = 'away_goals', y = 'season', ci= None, color ='red', estimator = sum );
plt.ylabel('Temporadas');
```

```python
Base_Dados['home_team'].unique()
```

```python
Base_Dados['Total_Goals'] = Base_Dados['home_goals'] + Base_Dados['away_goals']
```

```python
plt.figure(figsize =(14,5))
plt.title('Quantidade de Gols Premier League Temporadas 2006 a 2018', loc='left', fontsize = 14)
sns.barplot(data = Base_Dados, x = 'Total_Goals', y = 'season', ci= None, color ='#191970', estimator = sum );
plt.ylabel('Temporadas');
```

```python
plt.figure(figsize =(22,12))
plt.title('Quantidade de Gols Premier League Temporadas 2006 a 2018', loc='left', fontsize = 14)
sns.barplot(data = Base_Dados, x = 'Total_Goals', y = 'home_team', ci= None, color ='#07f23a', estimator = sum );
plt.ylabel('Temporadas');
```

```python
plt.figure(figsize =(22,12))
plt.title('Quantidade de Gols Premier League Temporadas 2006 a 2018', loc='left', fontsize = 14)
sns.barplot(data = Base_Dados, x = 'Total_Goals', y = 'away_team', ci= None, color ='orange', estimator = sum );
plt.ylabel('Temporadas');
```

```python
plt.figure(figsize =(22,12))
plot = sns.barplot(data=Base_Dados, x='Total_Goals', y='home_team', ci=False);
for i in plot.patches:
    plot.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
```
